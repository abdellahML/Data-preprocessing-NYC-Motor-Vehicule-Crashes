```python
#fill the zip_code to get the nearest zip_code in order to get the borough
df.sort_values(by='location',inplace=True)
df.zip_code.fillna(method='ffill', inplace=True)
df.reset_index(drop=True,inplace=True)
df
```

```python
df.zip_code = df.zip_code.astype(int)
```

```python
#df.dtypes
```

```python
#fill the zip_code based on the most used for a specific brough
df1 = df.groupby('borough')['zip_code'].apply(lambda x: x.value_counts().index[0]).reset_index()
df1
```

```python
#we can attribute the borough with the nearest Zip_code of the previous DataFrame (df1) 
    #or we can go back to the original and sorted DataFrame (df) to look for the nearest zip_code:
    
#attribute the borough
nan_index = df.loc[pd.isna(df["borough"]), :].index
# df.zip_code[nan_index[8000-10]]
```

```python
for i in nan_index:
    a = df.zip_code[nan_index[i-1]]
    b = df.zip_code[nan_index[i]]
    c = df.zip_code[nan_index[i+1]]
    if ( abs(a-b)< abs(b-c)):
        df.borough[i]=df.borough[i-1]
    else :
        df.borough[i]=df.borough[i+1]
```
